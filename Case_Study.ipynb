```python
import pandas as pd
import numpy as np
import datetime as dt

```

```python
demo = pd.read_csv("demo.csv")
open_accounts = pd.read_csv("open_accounts.csv")
transactions = pd.read_csv("transactions.csv")
calls_n = pd.read_csv("calls_north_center.csv")
calls_se = pd.read_csv("calls_south_east_center.csv")
```

```python
open_accounts["account_no"].dropna()
```

```python
#Q1 No Of Open Accounts

pd.pivot_table(open_accounts,index =["cust_id"],values ="account_no", aggfunc ='count' )
```

```python
calls_se["durationmins"] = calls_se["durationmins"]*60
calls_se.rename(columns = {'durationmins':'durationsecs'}, inplace = True)
calls_se['Flag'] = 1
calls_n['Flag'] = 0
calls = pd.concat([calls_n,calls_se])
```

```python
NEW_O_C = pd.merge(open_accounts,calls, how='inner', on="account_no", left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)
NEW_O_C
```

```python
#Q2 No Of Inbound Calls

df_in=NEW_O_C[NEW_O_C["call_type"]=="in"]
pd.pivot_table(df_in,index =["cust_id"],values ="call_type", aggfunc ='count' )
```

```python
#Q3 Closing Balance

filt_minus = (open_accounts["balance_type"]=="-") 

df1=open_accounts.loc[filt_minus, ["cust_id","balance"]]

df1['balance'] ='-' + df1['balance'].astype(str)

df2 = open_accounts.loc[~filt_minus, ["cust_id","balance"]]

cb1= df1.append(df2, ignore_index = True).set_index('cust_id').sort_index()

cb1.rename(columns = {"balance" : "closing_balance"}, inplace = True)

cb1['closing_balance'] = cb1['closing_balance'].astype(int)

cb1
```

```python
master1 = pd.merge(NEW_O_C,demo, how='inner', on="cust_id", left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)
    
```

```python
#Q4 Gender 
master1['gender'] = master1['gender'].str.replace('male', 'm')
master1['gender'] = master1['gender'].str.replace('ma', 'm')

q4 =master1[["cust_id",'gender']].set_index('cust_id')

q4

```

```python
#Q5 Marital Status 
master1[["cust_id",'marital_status']].drop_duplicates()
```

```python
#Q6 Average time spent by a customer on all calls with the bank 

pd.pivot_table(master1,index =["cust_id"],values ="durationsecs", aggfunc ='mean').round(2)
```

```python
#Q7 Number of days since last call was received/made from/to the customer

import datetime

master1["called_on"] = pd.to_datetime(master1["called_on"]) 

k1= pd.pivot_table(master1,index =["cust_id"],values ="called_on", aggfunc ='max' )

k1["current_date"] =datetime.date.today()

k1["current_date"] =pd.to_datetime(k1["current_date"]) 

k1['no_of_days'] = (k1['current_date'] - k1['called_on']) / np.timedelta64(1, 'D')

k1

```

```python
#Q8 Month on books

open_accounts["open_date"] = pd.to_datetime(open_accounts["open_date"]) 

m1= pd.pivot_table(open_accounts[open_accounts['open_date'].notna()],index =["cust_id"],values ="open_date", aggfunc ='min' )

m1["current_date"] =datetime.date.today()

m1["current_date"] =pd.to_datetime(m1["current_date"]) 

m1['months_booked'] = (m1['current_date'] - m1['open_date']) / np.timedelta64(1, 'M')

m1.round(2)
```

```python
#Q9  Age of the customer
master1["dob"] = pd.to_datetime(master1["dob"])

a1= pd.pivot_table(master1[master1['dob'].notna()],index =["cust_id"],values ="dob", aggfunc ='max' )

a1["current_date"] =datetime.date.today()

a1["current_date"] =pd.to_datetime(a1["current_date"]) 

a1['AGE'] = (a1['current_date'] - a1['dob']) / np.timedelta64(1, 'Y')

a1 = a1[(a1['AGE']>0)] 

a1['AGE'] = a1['AGE'].astype('int64')

a1
```

```python
#Q12  Opening balance of the saving account

# join open_accounts and tran



tab1 = pd.merge(open_accounts,transactions, how='right', on="account_no", left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)

tab1=tab1.drop_duplicates()

tab1 = tab1.dropna(how = "all", axis =1)

tab1["account_type"] = tab1["account_type"].replace("savings", "saving", regex=True)

filt_s = (tab1["account_type"]=="saving")



q12a = tab1.loc[filt_s, ['cust_id','account_no','tran_type','tran_amount']]

filt_credit = (q12a["tran_type"]=="credit") 

q12b=q12a.loc[filt_credit, ["cust_id","tran_amount","tran_type","account_no"]]

q12b['tran_amount'] = q12b['tran_amount'].apply(lambda x: x*-1)

q12c = q12b.drop_duplicates()

q12c

q12d =q12a.loc[~filt_credit, ["cust_id","tran_amount","tran_type","account_no"]]

q12d = q12d.drop_duplicates()


q12e= q12c.append(q12d, ignore_index = True).set_index('cust_id').sort_index()

q12e

q12tran=pd.pivot_table(q12e ,index =["cust_id"],values ="tran_amount", aggfunc ='sum' )

q12tran

q12f = tab1.loc[filt_s, ['cust_id','balance']]

q12f =pd.pivot_table(q12f ,index =["cust_id"],values ="balance", aggfunc ='sum' )

q12final = pd.merge(q12tran,q12f, how='outer', on="cust_id", left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)

q12final["opening_balance"] = q12final["tran_amount"] - q12final["balance"]

q12final
```

```python
#13 Opening balance of the current account
tab2 = pd.merge(open_accounts,transactions, how='left', on="account_no", left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)


filt_c = (tab2["account_type"]=="current")

q13a = tab2.loc[filt_c, ['cust_id','account_no','tran_type','tran_amount']]

filt_credit = (q13a["tran_type"]=="credit") 

q13b=q13a.loc[filt_credit, ["cust_id","tran_amount","tran_type","account_no"]]

q13b['tran_amount'] = q13b['tran_amount'].apply(lambda x: x*-1)

q13c = q13b.drop_duplicates()

q13c

q13d =q13a.loc[~filt_credit, ["cust_id","tran_amount","tran_type","account_no"]]

q13d= q13d.drop_duplicates()


q13e= q13c.append(q13d, ignore_index = True).set_index('cust_id').sort_index()

q13e
q13tran=pd.pivot_table(q13e ,index =["cust_id"],values ="tran_amount", aggfunc ='sum' )

q13tran
q13f = tab1.loc[filt_c, ['cust_id','balance']]

q13f =pd.pivot_table(q13f ,index =["cust_id"],values ="balance", aggfunc ='sum' )


q13final = pd.merge(q13tran,q13f, how='outer', on="cust_id", left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)

q13final.drop(q13final.tail(2).index,inplace=True)


q13final.at[67687, 'balance'] = 6768
q13final["opening_balance"] = q13final["tran_amount"] + q13final["balance"]

q13final



```
